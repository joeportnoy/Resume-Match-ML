```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("../resources/resume_data.csv")

# Display basic info and sample rows
# print(df.info())
df.head()
```

```python
# show column names
df.columns  
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.replace('﻿', '').str.lower()

# Fill NaNs with empty strings and convert to lowercase
df['career_objective'] = df['career_objective'].fillna('').str.lower()
df['job_position_name'] = df['job_position_name'].fillna('').str.lower()
df['responsibilities.1'] = df['responsibilities.1'].fillna('').str.lower()
df['degree_names'] = df['degree_names'].fillna('').str.lower()
df['major_field_of_studies'] = df['major_field_of_studies'].fillna('').str.lower()
df['experiencere_requirement'] = df['experiencere_requirement'].fillna('').str.lower()


# Convert 'skills' from string list to a flat lowercase string
import ast

def safe_parse_skills(x):
    try:
        skills_list = ast.literal_eval(x)
        if isinstance(skills_list, list):
            return ' '.join(skills_list).lower()
    except (ValueError, SyntaxError):
        return ''
    return ''

df['skills'] = df['skills'].fillna('').apply(safe_parse_skills)
```

```python
# Combine into a single searchable field
# df['combined_text'] = (
#     df['career_objective'] + ' ' +
#     df['job_position_name'] + ' ' +
#     df['responsibilities.1'] + ' ' +
#     df['degree_names'] + ' ' +
#     df['major_field_of_studies'] + ' ' +
#     df['experiencere_requirement'] + ' ' +
#     df['skills']
# )

df['combined_text'] = (
    df['career_objective']
)
```

```python
# Define keywords for filtering
keywords = [
    "data analyst", "data analytics", "business analyst", "data scientist",
    "data visualization", "analytics", "data analysis", "sql", "power bi",
    "tableau", "python", "r programming", "machine learning", "excel"
]
```

```python
# Filter the dataset
df_filtered = df[df['combined_text'].apply(lambda x: any(keyword in x for keyword in keywords))]
```

```python
# show filtered columns
df_filtered[['job_position_name', 'career_objective', 'skills', 'matched_score', 'degree_names', 'major_field_of_studies','experiencere_requirement']]
```

```python
import pandas as pd
import numpy as np
import ast

# Step 1: Load your dataset
# df_filtered = pd.read_csv('your_data.csv')

# Step 2: Replace empty strings and 'n/a' with NaN
df_filtered.replace(['', 'n/a', 'N/A', 'NA'], np.nan, inplace=True)

# Step 3: Safely convert string representations of lists
def safe_literal_eval(val):
    if isinstance(val, list):
        return val
    if isinstance(val, float) and pd.isna(val):  # Handles NaN
        return []
    try:
        result = ast.literal_eval(val)
        return result if isinstance(result, list) else [result]
    except (ValueError, SyntaxError):
        return [val.strip()] if isinstance(val, str) else []

# Step 4: Normalize job titles
df_filtered['job_position_name'] = df_filtered['job_position_name'].str.strip().str.title()

# Step 5: Fill missing values
df_filtered['career_objective'] = df_filtered['career_objective'].fillna('Not provided')
df_filtered['skills'] = df_filtered['skills'].fillna('Not provided')
df_filtered['experiencere_requirement'] = df_filtered['experiencere_requirement'].fillna('Not specified')

# Step 6: Filter by matched score
df_filtered = df_filtered[df_filtered['matched_score'] >= 0.4]

# Step 7: Keep only the required columns
df_filtered = df_filtered[['job_position_name', 'career_objective', 'skills', 'matched_score',
                           'degree_names', 'major_field_of_studies', 'experiencere_requirement']]

# Step 8: Reset index
df_filtered.reset_index(drop=True, inplace=True)

# View cleaned data
print(df_filtered.head())

```
